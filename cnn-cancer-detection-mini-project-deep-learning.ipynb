## > Exploratory Data Analysis (EDA)

Let's perform the initial analysis. The objective is understand the dataset and visualize sample images. After that we will check the distribution on the classes.

1. Load modules
   

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from glob import glob 
import matplotlib.pyplot as plt
import keras,cv2,os
from keras.layers import Dense, Dropout, Flatten, BatchNormalization, Activation, Input
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D
from tqdm.notebook import tqdm
from tqdm import trange
import os
#for dirname, _, filenames in os.walk('/kaggle/input'):
#    for filename in filenames:
#        print(os.path.join(dirname, filename))
print('Import complete')
```

2. Load files and labels

```python
train_p = '/kaggle/input/histopathologic-cancer-detection/train/'
#labels
labels = pd.read_csv("/kaggle/input/histopathologic-cancer-detection/train_labels.csv") 
#print(glob(os.path.join(train_p,'*.tif')))
# load .tif files
df = pd.DataFrame({'path': glob(os.path.join(train_p,'*.tif'))})

# get file filename on id
df['id'] = df.path.map(lambda x: x.split('/')[5].split(".")[0]) 
df = df.merge(labels, on = "id")
df.head(3)
print('Loaded!')
```

the full dataset contains around 220k. It will take a while to process even with GPUs. Let's create a variable to test our model with less samples (e.g. 1000) and before submission I can run it with all samples.

3. Import images (define small N to speedup)

```python
#define the total of samples to train, change to len(df) to train with the whole image set
#N=1000
N=len(df)

X = np.zeros([N,96,96,3],dtype=np.uint8) 
#loax X images
y = df['label'].to_numpy()[0:N]
for i, row in tqdm(df.iterrows(), total=N):
    if i == N:
        break
    X[i] = cv2.imread(row['path'])

print('df Images created')
```

```python
#plot images from dataset
fig = plt.figure(figsize=(10, 4), dpi=150)
for plotNr,idx in enumerate(np.random.randint(0,N,4)):
    ax = fig.add_subplot(1, 4, plotNr+1, xticks=[], yticks=[])
    plt.imshow(X[idx])
    ax.set_title('Label: ' + str(y[idx]))
```

After a quick sanity check, we can ensure the images and labels are loaded correctly. Label 1 indicates presence of cancer, label 0 no cancer.

4. Check distribution

```python
positive = X[y == 1]
negative = X[y == 0]
print("Positive: "+str(len(positive)))
print("Negative: "+str(len(negative)))
```

## > Create the model

```python
#split 80% 
split = int(np.round(0.8 * y.shape[0]))


np.random.seed(42)

#shuffle
idx = np.arange(y.shape[0])
np.random.shuffle(idx)
X = X[idx]
y = y[idx]
print('Suffle')
```

```python
kernel = (3,3)
pool= (2,2)
first_filters = 32
second_filters = 64
third_filters = 128
dropout_conv = 0.3
dropout_dense = 0.5

#model example mixed from other solutions available at Kaggle
#sequential model
model = Sequential([
    Input(shape=(96, 96, 3)),
    # 1st block
    Conv2D(first_filters, kernel),
    BatchNormalization(),
    Activation("relu"),
    Conv2D(first_filters, kernel, use_bias=False),
    BatchNormalization(),
    Activation("relu"),
    MaxPool2D(pool_size = pool), 
    Dropout(dropout_conv),
    # 2nd block
    Conv2D(second_filters, kernel, use_bias=False),
    BatchNormalization(),
    Activation("relu"),
    Conv2D(second_filters, kernel, use_bias=False),
    BatchNormalization(),
    Activation("relu"),
    MaxPool2D(pool_size = pool),
    Dropout(dropout_conv),
    # 3rd block
    Conv2D(third_filters, kernel, use_bias=False),
    BatchNormalization(),
    Activation("relu"),
    Conv2D(third_filters, kernel, use_bias=False),
    BatchNormalization(),
    Activation("relu"),
    MaxPool2D(pool_size = pool),
    Dropout(dropout_conv),
    # Dense layer
    Flatten(),
    Dense(256, use_bias=False),
    BatchNormalization(),
    Activation("relu"),
    Dropout(dropout_dense),
    # Sigmoid function
    Dense(1, activation = "sigmoid")
])

#compile the model
model.compile(loss=keras.losses.binary_crossentropy,
              optimizer=keras.optimizers.Adam(0.001), 
              metrics=['accuracy'])

print('Model created')
```

Train model

```python
#train the machine learning model
batch=50
for epoch in range(3):  # Train the model 3 times
    iterations = np.floor(split / batch).astype(int) # number of iterations
    loss,acc = 0,0 
    with trange(iterations) as t: 
        #inside the batch, train images
        for i in t:
            start_idx = i * batch 
            x_batch = X[start_idx:start_idx+batch] 
            y_batch = y[start_idx:start_idx+batch] 
            metrics = model.train_on_batch(x_batch, y_batch) 
            loss = loss + metrics[0] 
            acc = acc + metrics[1] 
            #display output
            t.set_description('Running training epoch ' + str(epoch)) 
            t.set_postfix(loss="%.2f" % round(loss / (i+1),2),acc="%.2f" % round(acc / (i+1),2)) 

print('Trainning complete!')
```

Verify the model

```python
iterations = np.floor((y.shape[0]-split) / batch).astype(int) #number of batches on validation set
loss,acc = 0,0
#validation loop
with trange(iterations) as t:  
    for i in t:
        start_idx = i * batch
        x_batch = X[start_idx:start_idx+batch] 
        y_batch = y[start_idx:start_idx+batch]

        #evaluate the model on this bach
        metrics = model.test_on_batch(x_batch, y_batch)
        
        loss = loss + metrics[0]
        acc = acc + metrics[1]
        t.set_description('Running training')
        t.set_description('Running validation')
        
        #update progess bar to show the running average loss and accuracy
        t.set_postfix(loss="%.2f" % round(loss / (i+1),2),acc="%.2f" % round(acc / (i+1),2))

#print final avg loss and accurary for whole validation
print("Validation loss:",loss / iterations)
print("Validation accuracy:",acc / iterations)
```

## > Submission

```python
print('Starting prediction!')
test_p = '/kaggle/input/histopathologic-cancer-detection/test/'
test_f = glob(os.path.join(test_p,'*.tif'))
submission_list = []
batch_size = 1000 

#process batches
for idx in tqdm(range(0, len(test_f), batch_size), desc="Processing batches"):
    batch_paths = test_f[idx:idx + batch_size]
    batch_images = [cv2.imread(path) for path in batch_paths]
    batch_df = pd.DataFrame({
        'path': batch_paths,
        'id': [path.split('/')[5].split(".")[0] for path in batch_paths]
    })
    
    K_test = np.stack(batch_images)
    batch_df['label'] = model.predict(K_test, verbose=0)
    submission_list.append(batch_df[["id", "label"]])
    
#create submission
submission = pd.concat(submission_list, ignore_index=True)
print("Processing complete.")
#export to file
submission.to_csv("submission.csv", index = False, header = True)
print("submission.csv saved")
```
